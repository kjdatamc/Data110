### 3. Multiple Lines

- Create a plot with the following three functions on the same graph:
 - $f(x) = x^3$
 - $g(x) = e^x$
 - $h(x) = \log ⁡(x)$ (for $x > 0$)
- Use different colors for each line
- Add a legend to distinguish between the functions


````python
x=np.linspace(-2,4)

f_3=pow(x,3)
plt.plot(x,f_3,color='#00f5f5',label='$f(x)=x^3$')

g_3=np.e**x
plt.plot(x,g_3,color='#00a5a5',label='$g(x)=e^x$')

x_clipped=np.clip(x, 0.001, None)
h_3=np.log(x_clipped)
plt.plot(x_clipped,h_3,color='#004545',label='$h(x)=log(x)$')

plt.legend()

plt.show()
````

### 8. Customizing Scatter Plots

- Generate data for the function $f(x) = x sin(x)$ over the range [ 0 , 4π ]
- Create a scatter plot of this data with a line connecting the points
- Customize the plot by changing the marker size, line width, marker face color, marker edge color, and marker edge width


````python
x_8=np.linspace(0,4*np.pi)

f_8=x_8*np.sin(x_8)
plt.plot(x_8,f_8,'-o', color='k',
         markersize=10, linewidth=2,
         markerfacecolor='#ee99ee',
         markeredgecolor='k',
         markeredgewidth=2)

plt.title(" $f(x) = x sin(x)$ ")

plt.show()
````

#### Notes on my selected functions:

I didn't limit the number of points plotted because it negatively impacted the appearance of the $tan(x)$ function when I tested using 17 and 33 points.
````python
x_10=np.linspace(-1*np.pi,1*np.pi)

f_10=np.tan(x_10)
plt.plot(x_10,f_10,':oc',
         markersize=5, linewidth=2,
         markerfacecolor='c',
         markeredgecolor='c',
         markeredgewidth=1,
         label='$tan(x)$')

i_10=np.arctan(x_10)
plt.plot(x_10,i_10,'-^y',
         markersize=5, linewidth=1,
         markerfacecolor='y',
         markeredgecolor='y',
         markeredgewidth=1,
         label='$arctan(x)$')

plt.ylim(-10,10)

plt.xlabel('x')
plt.ylabel('y')
plt.title(" Comparison of TAN and ARCTAN ", fontsize=18)

plt.legend();
````

### Analysis of Different Chart Styles

I personally gravitate toward the `seaborn whitegrid` style because of the modern, minimalist look. In this case, I also think the the gridlines of the `seaborn whitegrid` style gives a better sense of the expanding oscillation of the function. However, the bright white gridlines of the `dark background` style are a bit harsh. In my opinion they're actually distracting from the plotted function because the gridlines are competing too much with the function for attention.

I do notice that once I import a style, certain elements of that style remain even when I replace it with another. I assume this is because some style elements are defined in one but not in the other - essentially those undefined elements in the new style continue to use the defined elements of the replaced style. For example, the `dark background` seems to use a light cyan/teal line for the plotted function, which stays even when I switch back to `seaborn whitegrid`. Exercise #2 shows a blue line when using `seaborn whitegrid`, which is the same as the default. That suggests the `seaborn whitegrid` styling focuses on the plot background and grid rather than the plotted elements.
