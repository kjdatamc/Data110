## Task 2: Change in Crime Rates (2022 vs. 2023) with Unique Colors

In this task, you will create a **horizontal bar chart** that visualizes the **percentage change** in offense rates from 2022 to 2023 for each crime type. Each bar should have a unique color for better distinction.

- Create a horizontal bar chart showing the **change in offense rates**.
- Assign a unique color to each crime type.
- Label the axes and add a title.

```python
Offense=['Aggravated Assault','Burglary','Larceny','Motor Vehicle Theft','Murder','Rape (revised definition)','Robbery']
# Rate_2023=[264.1,250.7,1347.2,318.7,5.7,38.0,66.5]
Rate_Change_22_23=[-3.3,-8.1,-4.9,12.0,-12.0,-9.8,-0.8]

plt.figure(figsize=(12, 7))

#assign unique color to each crime type
rainbow = ['#D1B1CB', '#444E70', '#007991', '#439A86', '#E9D985', '#F7A072', '#F45B69']

plt.barh(Offense, Rate_Change_22_23, color=rainbow, zorder=2)

#label axes, add title
plt.ylabel('Offense')
plt.xlabel('Percent Change (2022-2023)')
plt.title('Change in Offense Rates (2022-2023)')

plt.grid(axis='x', zorder=0);
```

   - Explain why seeing the **change** over time is more insightful than just seeing the total number of offenses for a single year.


### **My Response**

#### Absolute Numbers vs Percentages

When comparing years for multiple types of offenses, it's helpful to look at the **percentage change** rather than the **absolute numbers** because it gives a better sense of the ***context***.

For example, in absolute numbers, the murder rate decreased by 0.8 between 2022 and 2023. That seems like a very small change compared to a reduction in the larceny rate of 69.4. However, there are more than 200 times as many larceny offenses as there are murders. That means a change in the absolute number of offenses represents a much larger percentage change for murder than for larceny. In fact, the murder rate decreased by a huge 12%, but the larceny rate only decreased by 4.9%.

#### Change Over Time

Task 1 shows the crime rates for a single year, which is a snapshot of a point in time that doesn't provide the ***context*** of what the trend in crime rates is. We can't tell from that graph if crime is increasing, decreasing, or steady. We can't tell if certain types of offenses are not following the trend that the others are. Without that context, we can't determine if there are problems or successes with our approach to crime reduction & prevention.

In this case, most crime is decreasing, but motor vehicle theft has significantly increased. It would be important to do further investigation to determine why motor vehicle theft is not following the trend that other types of offenses are.

```python
# Sort by change in crime rate, adjust axes labels, resize tick marks

data = {'Offense' : ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (revised definition)', 'Robbery'],
        'Percent Change' : [-3.3,-8.1,-4.9,12.0,-12.0,-9.8,-0.8]}

dm = pd.DataFrame(data)
rainbow = ['#D1B1CB', '#444E70', '#007991', '#439A86', '#E9D985', '#F7A072', '#F45B69']

dm_sorted = dm.sort_values(by='Percent Change', ascending=True)

plt.figure(figsize=(12, 7))

plt.barh(dm_sorted['Offense'], dm_sorted['Percent Change'], color=rainbow, zorder=2)

plt.ylabel('Offense')
plt.xlabel('Percent Change in Offenses per 100,000 People (2022-2023)')
plt.title('Change in Crime Rates (2022-2023)')

ax = plt.gca()
ax.tick_params(axis='x', labelsize=8)
ax.tick_params(axis='y', labelsize=8)
plt.grid(axis='x', linestyle='-', color='#dddddd', zorder=0)

plt.show()
```

## Optional Task: Combined Bar Chart for Crime Rates in 2022 and 2023

In this optional task, you will create a **combined bar chart** that shows the crime rates for both **2022 and 2023**.

- Calculate the **crime rates for 2022** based on the 2023 values and the percentage changes provided.
- Create a **combined bar chart** that compares the crime rates for 2022 and 2023 for each offense type.
- Use different colors to distinguish between the years 2022 and 2023.
- Label the axes and add a title.
- Add a legend to indicate which bars represent which year.

```python
# Calculate 2022 rates and create 22-23 bar chart, unsorted

Offense=['Aggravated Assault','Burglary','Larceny','Motor Vehicle Theft','Murder','Rape (revised definition)','Robbery']
Offense_index=np.arange(len(Offense))

Rate_2023=[264.1,250.7,1347.2,318.7,5.7,38.0,66.5]
Rate_Change_22_23=[-3.3,-8.1,-4.9,12.0,-12.0,-9.8,-0.8]
Rate_2022=[a/((b/100)+1) for a,b in zip(Rate_2023,Rate_Change_22_23)]

plt.figure(figsize=(12, 6))

plt.barh(Offense_index+.18,Rate_2022,height=.36,label='2022',color='#444E70', zorder=2)
plt.barh(Offense_index-.18,Rate_2023,height=.36,label='2023',color='#439A86', zorder=2)
plt.yticks(Offense_index,Offense)
plt.legend()

plt.ylabel('Offense')
plt.xlabel('Offenses per 100,000')
plt.title('Crime Rates (2022-2023)')

ax = plt.gca()
ax.tick_params(axis='x', labelsize=8)
ax.tick_params(axis='y', labelsize=8)
plt.grid(axis='x', linestyle='-', color='#dddddd', zorder=0)

plt.show()
```
